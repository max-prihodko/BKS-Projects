```python
import pandas as pd
from datetime import datetime, timedelta
import time
```

```python
df1 = pd.read_excel('Табл1.xlsx')
df2 = pd.read_excel('Табл2.xlsx')
df3 = pd.read_excel('Табл3.xlsx')
#df4 = pd.read_excel('Табл4.xlsx')
```

```python
df_1 = df1[df1['Страховая_Регион'] == df1['Регион']]
df_2 = df2[df2['Страховая_Регион'] == df2['Регион']]
df_3 = df3[df3['Страховая_Регион'] == df3['Регион']]


df_region = pd.merge(df_1,df_2,  on = ['Страховая_Регион','Регион'], how = 'inner')
df_region = pd.merge(df_region, df_3,  on = ['Страховая_Регион','Регион'], how = 'inner')

df_region.to_excel('region_2019.xlsx')
df_region.head(20)
```

```python
df_1_new = df1[df1['Страховая_Регион'] != df1['Регион']]
df_2_new = df2[df2['Страховая_Регион'] != df2['Регион']]
df_3_new = df3[df3['Страховая_Регион'] != df3['Регион']]

df = pd.merge(df_1_new,df_2_new,  on = ['Страховая_Регион','Регион'], how = 'inner')
df = pd.merge(df,df_3_new,  on = ['Страховая_Регион','Регион'], how = 'inner')
df[df['Страховые премии 2019'] != 0].shape[0]
```

# Подсчёт средней выплаты и средней премии по страховым за 2019
- загружаем 2 таблицы: 

1) по общему размеру выплат по ОСАГО в разрезе страховщиков

2) по количеству выплат по ОСАГО в разрезе страховщиков

```python
df = df[df['Выплаты 2019'] != 0].reset_index(drop=True)
df.rename(columns = {'Страховая_Регион' : 'Страховая'}, inplace = True)
df['Средняя премия 2019'] = df['Страховые премии 2019'] / df['Заключенные договора 2019']
df
```

```python
df_1 = pd.read_excel('Табл4.xlsx')
df_2 = pd.read_excel('Табл5.xlsx')

df_new = pd.merge(df_1,df_2,  on = ['Страховая'], how = 'inner')
df_new['Средняя выплата 2019'] = df_new['Выплаты 2019'] / df_new['Кол-во выплат 2019']
df_new = df_new.drop(['Кол-во выплат 2019'], axis = 1)
df_new = df_new.drop(['Выплаты 2019'], axis = 1)
df_new
```

```python
df = pd.merge(df,df_new,  on = ['Страховая'], how = 'inner')
df = df[df['Страховые премии 2019'] > 0].reset_index(drop=True)
df.to_excel('CB_OSAGO_2019.xlsx')
df
```
